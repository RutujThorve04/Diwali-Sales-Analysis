```python
df['Amount']=df.groupby('Product_Category')['Amount'].transform(lambda x:x.fillna(x.mean()))
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.rename(columns={"Marital_Status" :'Married'})
```

```python
def categories_age(age):
    if age<=17:
        return'0-17'
    if age<=25:
        return '18-25'
    if age<=35:
        return'26-35'
    if age<=45:
        return'36-45'
    if age<=55:
        return'46-55'
    else:
        return '55+'
```

```python
df['Age_Group']=df['Age'].apply(categories_age)
```

```python
df.drop('Age',axis=1,inplace=True)
```

# Gender

# AGE and GENDER

```python
sns.countplot(data=df,x='Age_Group',hue="Gender")
```

```python
sales_age=df.groupby(['Age_Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x="Age_Group",y="Amount",data=sales_age)
```

# STATE

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sns.barplot(x="State",y="Orders",data=sales_state)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

# MARITAL STATUS

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

# OCCUPATION

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
